## **Nigeria Housing Price Prediction**
### **Introduction**

The aim of this project is to apply linear regression techniques to a real-world Nigerian real estate dataset to predict house prices. The goal is to practice and reinforce core machine learning concepts such as ordinary least squares (OLS), regularization with Ridge regression, handling multicollinearity, and evaluating model performance. The dataset provides a realistic scenario for predictive modeling in the Nigerian housing market.

The dataset was obtained from Kaggle, contributed by [chik0di](http://kaggle.com/chik0di). It includes a variety of attributes describing each property, offering valuable insights into the dynamics of the Nigerian housing market. Researchers, data analysts, and developers can use this dataset for machine learning projects, pricing models, or market studies.

```python
# import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import our dataset into the pandas dataframe

df = pd.read_csv('crib.csv')
df.head()
```

```python
# before going on to remove the columns that are not needed let us also check the data types of the column

df.info()
```

### **Data Preprocessing / Issues**

1. **Drop irrelevant columns**  
   - Columns like `Unnamed: 0` and `Property Ref` do not add predictive value and will be removed.  
   - `Service Charge` columns contain only 8 non-null values out of 1,668 observations (~0.5%), making them unreliable. These will be dropped.
     

2. **Convert date columns**  
   - `Added on` and `Last Updated` will be converted to `datetime` format to allow for date-based analysis or feature engineering.
     

3. **Handle columns with missing values but predictive importance**  
   - The `Servicing` (139 non-null) and `Furnishing` (140 non-null) columns contain a large number of missing values but are important for predicting house prices.  
   - To avoid making strong assumptions about the meaning of missing values, these columns were filled with `Unknown`.

    
4. **Feature engineering**  
   - The `Added On` and `Last Updated` datetime columns were used to create a new feature, `Listing_Duration_Days`, which represents the number of days between when a property was added to the market and its most recent update.  
   - This feature captures how long a listing has been active, which may influence property pricing.




```python
# drop the unnamed column, the property ref column and the Service Charge column

columns_to_drop = ['Unnamed: 0', 'Property Ref', 'Service Charge']

for col in columns_to_drop:
    df.drop(col, axis=1, inplace=True)

#use df.info() to check the columns
df.info()
```

```python
# issue 2 (convert the date columns in string to date)

df['Added On'] = pd.to_datetime(df['Added On'])
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

# check datatypes if it was successful

df.info()
```

```python
df['Total Area'] = df['Total Area'].str.replace('sqm', '', regex=False).str.strip()
df['Total Area'] = pd.to_numeric(df['Total Area'], errors='coerce')

df['Covered Area'] = df['Covered Area'].str.replace('sqm', '', regex=False).str.strip()
df['Covered Area'] = pd.to_numeric(df['Covered Area'], errors='coerce')
```

```python
# fill the missing values with zero since 0 would not affect the model
numeric_cols = ['Bedrooms','Bathrooms','Toilets','Parking Spaces','Total Area','Covered Area']
df[numeric_cols] = df[numeric_cols].fillna(0)

#check the dataframe again
df.info()
```

```python
# let us check the servicing and furnishing columns

df[['Servicing', 'Furnishing']].value_counts()
```

```python
# let us input the missing values in servicing and furnishing columns as unknown to be safe

categorical_cols = ['Servicing','Furnishing']
df[categorical_cols] = df[categorical_cols].fillna('Unknown')
```

```python
# Simple Feature Engineering
df['Listing_Duration_Days'] = (df['Last Updated'] - df['Added On']).dt.days
df['Listing_Duration_Days'] = pd.to_numeric(df['Listing_Duration_Days'])
```

### **Exploratory Data Analysis**

```python
# now let us look into the target variable price

sns.histplot(df['Price'], bins = 50)
plt.title("Price Distribution of Houses in Nigeria")
plt.show()
```

*we can see from the plot above the our target varaible is skewed to the right, which can cause our model to perform poorly. So, to mitigate against this, we would perform log transformation on the price variable. This helps us see the distribution well and also helps the model because linear model assumes normally distributed errors*

```python
# perform log transformation and create a hist plot of the log transformed price
price_log = np.log1p(df['Price'])
sns.histplot(price_log, bins = 40)
plt.title("Price Distribution of Houses in Nigeria")
plt.show()
```

```python
numeric_df = df[numeric_cols]
numeric_df_list = list(numeric_df.columns)
```

```python
numeric_df_list
```

```python
fig, axs = plt.subplots(2, 3, figsize=(20, 6))
fig.subplots_adjust(hspace = 0.5, wspace = .2)
axs = axs.flatten()  # convert 2D array to 1D for easy indexing

for index, column in enumerate(numeric_df_list):
    axs[index].set_title("{} vs Price".format(column), fontsize=13)
    axs[index].scatter(df[column], np.log1p(df['Price']))
    axs[index].set_xlabel(column)
    axs[index].set_ylabel('Price')
```

*Exploratory analysis reveals weak but positive relationships between numeric features and price, along with noticeable skewness and outliers. To address this, we apply a log transformation to the target and use regularized regression models. Additionally, we create binary indicator columns for missing values, allowing the model to account for missingness and improve generalizability*
