## **Nigeria Housing Price Prediction Using Regularized Linear Model**
### **Introduction**

The aim of this project is to apply linear regression techniques to a real-world Nigerian real estate dataset to predict house prices. The goal is to practice and reinforce core machine learning concepts such as ordinary least squares (OLS), regularization with Ridge regression, handling multicollinearity, and evaluating model performance. The dataset provides a realistic scenario for predictive modeling in the Nigerian housing market.

### **About the Dataset**

The dataset was obtained from Kaggle, contributed by [chik0di](http://kaggle.com/chik0di). It contains information scraped from [Nigeria Property Centre](https://nigeriapropertycentre.com/for-sale/houses/showtype), providing detailed data on houses listed in the Nigerian real estate market, including housing trends and property pricing.

It includes a variety of attributes describing each property, offering valuable insights into the dynamics of the Nigerian housing market. Researchers, data analysts, and developers can use this dataset for machine learning projects, pricing models, or market studies.

### **Observations & Issues**

1. **Drop irrelevant columns**  
   - Columns like `Unnamed: 0` and `Property Ref` do not add predictive value and will be removed.  
   - `Service Charge` columns contain only 8 non-null values out of 1,668 observations (~0.5%), making them unreliable. These will be dropped.
     

2. **Convert date columns**  
   - `Added on` and `Last Updated` will be converted to `datetime` format to allow for date-based analysis or feature engineering.
     

3. **Handle columns with missing values but predictive importance**  
   - The `Servicing` (139 non-null) and `Furnishing` (140 non-null) columns contain a large number of missing values but are important for predicting house prices.  
   - To avoid making strong assumptions about the meaning of missing values, these columns were filled with `Unknown`.

    
4. **Feature engineering**  
   - The `Added On` and `Last Updated` datetime columns will be used to create a new feature, `Listing_Duration_Days`, which represents the number of days between when a property was added to the market and its most recent update.  
   - This feature captures how long a listing has been active, which may influence property pricing.
   - Add missingness indicators columns so the model can distinguish missing values from real zeros and improve generalizability.




```python
# drop the unnamed column, the property ref column and the Service Charge column

columns_to_drop = ['Unnamed: 0', 'Property Ref', 'Service Charge']

for col in columns_to_drop:
    df.drop(col, axis=1, inplace=True)

#use df.info() to check the columns
df.info()
```

```python
# issue 2 (convert the date columns in string to date)

df['Added On'] = pd.to_datetime(df['Added On'])
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

# check datatypes if it was successful

df.info()
```

```python
df['Total Area'] = df['Total Area'].str.replace('sqm', '', regex=False).str.strip()
df['Total Area'] = pd.to_numeric(df['Total Area'], errors='coerce')

df['Covered Area'] = df['Covered Area'].str.replace('sqm', '', regex=False).str.strip()
df['Covered Area'] = pd.to_numeric(df['Covered Area'], errors='coerce')
```

```python
# fill missing values with zero and add missingness indicators
# so the model can distinguish missing values from real zeros
numeric_cols = ['Bedrooms','Bathrooms','Toilets','Parking Spaces','Total Area','Covered Area']
for col in numeric_cols:
    df[col + '_missing'] = df[col].isna().astype(int)
    df[col] = df[col].fillna(0)

#check the dataframe again
df.info()
```

```python
# let us check the servicing and furnishing columns

df[['Servicing', 'Furnishing']].value_counts()
```

```python
# let us input the missing values in servicing and furnishing columns as unknown to be safe

categorical_cols = ['Servicing','Furnishing']
df[categorical_cols] = df[categorical_cols].fillna('Unknown')
```

```python
# Simple Feature Engineering
df['Listing_Duration_Days'] = (df['Last Updated'] - df['Added On']).dt.days
```

### **Data Preprocessing**

```python
# first we would train the model only on the numerical columns
numeric = ['Bedrooms_missing', 'Bathrooms_missing', 'Toilets_missing', 'Parking Spaces_missing', 
 'Total Area_missing', 'Covered Area_missing']
df_numeric = df[numeric_df_list + numeric + ['Price']] 
df_numeric.info()
```

```python
def train_val_test(df):
    n_test = int(len(df) *0.2)
    n_val = int(len(df) * 0.2)
    n_train = len(df) - n_test - n_val
    
    # we need to randomise the selection
    idx = np.arange(len(df))
    np.random.seed(2)
    np.random.shuffle(idx)
    
    df_train = df.iloc[idx[:n_train]]
    df_val = df.iloc[idx[n_train:n_train+n_val]]
    df_test = df.iloc[idx[n_train+n_val:]]
    
    df_train = df_train.reset_index(drop=True)
    df_val = df_val.reset_index(drop=True)
    df_test = df_test.reset_index(drop=True)

    y_train = np.log1p(df_train.Price.values)
    y_val = np.log1p(df_val.Price.values)
    y_test = np.log1p(df_test.Price.values)
    
    del df_train['Price']
    del df_val['Price']
    del df_test['Price']

    return df_train, df_test, df_val, y_train, y_val, y_test
```

```python
df_train, df_test, df_val, y_train, y_val, y_test = train_val_test(df_numeric)
```

```python
len(df_train), len(df_val), len(df_test)
```

### **Training linear Regression**

```python
def train_linear_regression_Reg(df, y, r):
    X = df.values
    ones = np.ones(len(X))
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX = XTX + r * np.eye(XTX.shape[0])
    XTX_inv = np.linalg.inv(XTX)
    XTy = X.T.dot(y)
    w = XTX_inv.dot(XTy)
    return w[0], w[1:]
```

```python
w0, w = train_linear_regression_Reg(df_train, y_train, 0.0001)
```

```python
X_train = df_train.values
y_pred = w0 + X_train.dot(w)
```

```python
def rmse(y, y_pred):
    se = (y - y_pred) **2
    mse = se.mean()
    return np.sqrt(mse)
```

```python
rmse(y_train, y_pred)
```

### **Tuning The Model**

```python
def model_tuning(df, y, alpha):
    x_val = df_val.values
    for r in alpha:
        w0, w = train_linear_regression_Reg(df, y, r)
        y_pred = w0 + x_val.dot(w)
        score = rmse(y_val, y_pred)
        print(f"bias: {w0}, alpha: {r}, score: {score}")
```

```python
model_tuning(df_train, y_train, [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10])
```

### **Retraining The Model**

```python
df_train_full = pd.concat([df_train, df_val])
```

```python
y_train_full = np.concatenate((y_train, y_val))
```

```python
w0, w = train_linear_regression_Reg(df_train_full, y_train_full, 0.0001)
```

```python
X_train = df_train_full.values
y_pred = w0 + X_train.dot(w)
```

```python
score = rmse(y_train_full, y_pred)
print(f"Train RMSE: {score}")
```

```python
# predict the Test RMSE using the Retrained model

x_test = df_test.values
y_pred = w0 + x_test.dot(w)
score = rmse(y_test, y_pred)
print(f"Test RMSE: {score}")
```

```python
sns.histplot(y_pred, color = 'red', alpha = 0.5, bins = 50)
sns.histplot(y_train_full, color = 'blue', alpha = 0.5, bins = 50)
plt.title('House Predicted Price vs House Price')
plt.show()
```

### **Training Both The Categorical and Numeric Columns**

```python
w0, w = train_linear_regression_Reg(df_train, y_train, 0.0001)
X_train = df_train.values
y_pred = w0 + X_train.dot(w)
score = rmse(y_train, y_pred).round(5)
print(f"Train RMSE: {score}")
```

### **Tuning The new Model**

```python
def model_tuning(df, y, alpha):
    x_val = df_val.values
    for r in alpha:
        w0, w = train_linear_regression_Reg(df, y, r)
        y_pred = w0 + x_val.dot(w)
        score = rmse(y_val, y_pred)
        print(f"bias: {w0}, alpha: {r}, score: {score}")
        
model_tuning(df_train, y_train, [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10])
```

### **Retraining the new model**

```python
# retrain the model using the alpha value with the best prediction on both the training and validation dataset
df_train_full = pd.concat([df_train, df_val])
y_train_full = np.concatenate((y_train, y_val))
w0, w = train_linear_regression_Reg(df_train_full, y_train_full, 0.0001)
X_train = df_train_full.values
y_pred = w0 + X_train.dot(w)
score = rmse(y_train_full, y_pred).round(5)
print(f"Retrained RMSE: {score}")
```

```python
# predict the test dataset using paramters gotten from training dataset
X_test = df_test.values
y_pred = w0 + X_test.dot(w)
score = rmse(y_test, y_pred).round(5)
print(f" Test RMSE: {score}")
```
